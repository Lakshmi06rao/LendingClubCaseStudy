```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
ld = pd.read_csv('C:/Users/hp/Downloads/loan/loan.csv')
ld
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# to check if any of the rows or columns have null (no data)
ld.isnull()
```

```python
#Drops dupilcate data
ld.drop_duplicates()
```

```python
#no duplicate data found.
#Deleting columns having null for all rows, before deletion total number of columns: 111
ld_cleaned = ld.dropna(axis=1, how='all')
ld_cleaned
```

```python
#fetch date and month and year where applicable for the follwoing date fields

def convert_date(date_str):
    return pd.to_datetime(date_str, format='%b-%y')

ld_cleaned['issue_d'] = ld_cleaned['issue_d'].apply(convert_date)
ld_cleaned['earliest_cr_line'] = ld_cleaned['earliest_cr_line'].apply(convert_date)

ld_cleaned['issue_month'] = ld_cleaned['issue_d'].dt.month
ld_cleaned['issue_year'] = ld_cleaned['issue_d'].dt.year

ld_cleaned['earliest_cr_line_month'] = ld_cleaned['earliest_cr_line'].dt.month
ld_cleaned['earliest_cr_line_year'] = ld_cleaned['earliest_cr_line'].dt.year
```

```python
## data being cleaned of special characters and setting their type to int or float
ld_cleaned['term'] = ld_cleaned['term'].apply(lambda x: int(x.replace(' months', '')))
ld_cleaned['int_rate'] = ld_cleaned['int_rate'].apply(lambda x: float(x.replace('%', '')))
ld_cleaned['zip_code'] = ld_cleaned['zip_code'].apply(lambda x: int(x.replace('xx', '')))
```

```python
#Remove columns that has maximum of null values
ld_cleaned = ld_cleaned.drop('acc_now_delinq', axis=1)
ld_cleaned = ld_cleaned.drop('delinq_amnt', axis=1)
ld_cleaned = ld_cleaned.drop('next_pymnt_d', axis=1)
ld_cleaned = ld_cleaned.drop('mths_since_last_delinq', axis=1)
ld_cleaned = ld_cleaned.drop('mths_since_last_record', axis=1)
# ld_cleaned = ld_cleaned.drop([]'acc_now_delinq', 'delinq_amnt', 'next_pymnt_d', 'mths_since_last_delinq' ,'mths_since_last_record'], axis=1, inplace=True)
```
